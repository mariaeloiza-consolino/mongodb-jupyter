2. Consultar atendimentos com valor entre 90 e 200 reais.


```python
dados_mongo = list(colecao.find ({
    'valor': {'$gte' : 90, '$lte' : 200}
}))

display(dados_mongo)
```

3. Consultar atendimentos das cidades Assis, Ourinhos, Pompeia e Marília.



```python
dados_cidade = list(colecao.find({
    'cidade' : {'$in': ['Ourinhos', 'Assis', 'Pompeia', 'Marilia']}
}))

display (dados_cidade)
```

4. Converter os resultados para DataFrame.


```python
df = pd.DataFrame(dados_cidade)
df
```

5. Verificar valores nulos.


```python
df.isnull().sum()
```

6. Calcular faturamento total.


```python
faturamento = df['valor'].sum()
print("Faturamento total: " , faturamento)
```

7. Calcular faturamento por cidade.


```python
faturamento_cidade = df.groupby('cidade')['valor'].sum().sort_values(ascending=False)
faturamento_cidade
```

8. Calcular faturamento por serviço.


```python
faturamento_servico = df.groupby('servico')['valor'].sum().sort_values(ascending=False)
faturamento_servico
```

9. Calcular ticket médio geral.


```python
ticket_medio_geral = df['valor'].mean()
print(f"Média do valor das consultas: {ticket_medio_geral: .2f}")
```

10. Calcular ticket médio por cidade.


```python
ticket_medio_por_cidade = df.groupby('cidade')['valor'].mean().round(2)
print(f"Média do valor das consultas por cidade: { ticket_medio_por_cidade}")
```

11. Identificar o serviço mais lucrativo.


```python
servico_mais_lucrativo = df.groupby('servico')['valor'].sum().idxmax()
servico_mais_lucrativo
```

12. Identificar a cidade com maior faturamento.

```python
cidade_maior_faturamento = df['cidade'].value_counts().idxmax()
cidade_maior_faturamento
```

13. Identificar os 3 maiores atendimentos.


```python
top_3_atendimentos = df.nlargest(3, 'valor')[['servico', 'valor']]
top_3_atendimentos
```
